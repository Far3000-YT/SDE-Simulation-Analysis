We defined the price model in src/sde_simulator/simulators.py, now we import it and use it to plot (we can change parameters to get different price simulations patterns... etc, like starting price, risk, average yearly return or anything !)

```python
import numpy as np
import matplotlib.pyplot as plt
import sys, os

notebook_dir = os.path.dirname(os.path.abspath("__file__"))
project_root = os.path.abspath(os.path.join(notebook_dir, '..'))
src_path = os.path.join(project_root, 'src')
if src_path not in sys.path:
    sys.path.append(src_path)

#importing simulation function here

from sde_simulator.simulators import simulate_gbm_em

S0 = 20    #initial stock price
MU = 0.1      #drift (10% annual return)
SIGMA = 0.2   #volatility (20% annual volatility)
T = 3          #time horizon (3 years)
DT = 1/252     #time step (daily steps, 1/252 assuming 252 trading days basically)

#run here
time_points, asset_path = simulate_gbm_em(S0, MU, SIGMA, T, DT)

#plot, basic matplotlib
plt.figure(figsize=(10, 6))
plt.plot(time_points, asset_path, color = "black")
plt.title(f'Geometric Brownian Motion Simulation (Euler Maruyama)')

plt.xlabel('Time (Years)') ; plt.ylabel('Asset Price ($)')
#grid, legend and then show graph
plt.grid(True); 
plt.show() 

print(f"Simulation finished :)")
print(f"Initial: {asset_path[0]:.2f}")
print(f"Final:   {asset_path[-1]:.2f}")
print(f"Steps:   {len(asset_path) - 1}")
```
